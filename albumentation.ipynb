```python
def myFig(img, label) :
    # Create figure and axes
    fig, ax = plt.subplots()

    # Display the image
    ax.imshow(img)

    if len(label)>0 :
        for _ in range(len(label)):
            
            # Create a Rectangle patch
            dw = img.shape[1]
            dh = img.shape[0]

            x1 = (label[_][0] - label[_][2] / 2) * dw
            y1 = (label[_][1] - label[_][3] / 2) * dh
            w = label[_][2] * dw
            h = label[_][3] * dh
            rect = patches.Rectangle((x1, y1), w, h, linewidth=1, edgecolor='r', facecolor='none')

            # Add the patch to the Axes
            ax.add_patch(rect)

    plt.show()
```

```python
k=0

for i in range(len(image_list)): # 모든 이미지를 augmentation
   # k += 1
    img_path = image_list[i]
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    bboxes = np.loadtxt(fname=label_list[i], delimiter=" ",ndmin=2)
    bboxes = np.roll(bboxes, 4, axis=1).tolist()
    
    for j in range(2): # 이미지 한장당 augmentation갯수
        k +=1
        transform = A.Compose([
            A.RandomRotate90(),
            A.Transpose(),

            A.OneOf([
                A.augmentations.transforms.GaussNoise(),
                A.GaussNoise(),
            ], p=.2),

            A.ShiftScaleRotate(shift_limit=.0625, scale_limit=.5,rotate_limit=45, p=.75),

            A.OneOf([
                A.MotionBlur(p=.2),
                A.MedianBlur(blur_limit=3,p=.1),
                A.Blur(blur_limit=3,p=.1)
            ], p=.2),

            A.OneOf([
                A.CLAHE(clip_limit=2),
                A.augmentations.transforms.Sharpen (),
                A.augmentations.transforms.Emboss(),
                A.RandomBrightnessContrast(),], p=.3),
            A.HueSaturationValue(p=.3),
        ],bbox_params=A.BboxParams(format="yolo", min_visibility=0.4, label_fields=[],))

        transformed = transform(image=image, bboxes=bboxes)
        transformed_image = transformed['image'] #<= save
        transformed_bboxes = transformed['bboxes']

        print(transformed_image.shape)
        print(transformed_bboxes) # <= save
        myFig(transformed_image,transformed_bboxes)

        filename='new'+'_'+str(k)

        im = Image.fromarray(transformed_image,'RGB')
        path_image='output/images/' # 이미지 결과 저장 경로
        im.save(path_image+filename+'.jpg')

        path_labels='output/labels/' # 라벨 결과 저장 경로
        file = open(path_labels+filename+'.txt', "w")
        l = ''
        for i in transformed_bboxes:
            for n,j in enumerate(i):
                l += str(j)+' '
            l += '\n'

        file.write(l)
        file.close()
```
