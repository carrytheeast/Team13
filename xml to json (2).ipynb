```python
import json
import xmltodict
import os 
import re 

total_main = []
total_sub = []
path = 'book_280n_pen_360n/annotations/val/'
box_id=0

for id,filename in enumerate(os.listdir(path)):
    id+=1
    with open(path+filename, encoding='utf-8') as f:
        doc = xmltodict.parse(f.read())
        json_data = json.loads(json.dumps(doc))
        print(filename)
        # main data(file name, height, width, id(image))---------------------------------------- 
        file_name= re.findall("\w+",filename)[0]
        height=json_data['annotation']['size']['height']
        width=json_data['annotation']['size']['width']
        id= id
        main = {"file_name":file_name, "height":height, "width":width, "id":id}
    
        total_main.append(main)
        
        # sub data(area, iscrowd, image_id, bbox, category_id, id(box), ignore, segmentation)----------------------------------------
        
        area=int(json_data['annotation']['size']['width'])*int(json_data['annotation']['size']['height'])
        iscrowd=0
        ignore= 0
        segmentation= []
        
        # object개수에 따라 bbox의 x y 구하는 식이 달라짐
        # object가 1개 json_data['annotation']['object']['bndbox']['xmin']
        # object가 2개이상 json_data['annotation']['object'][index]['bndbox']['xmin']
        
        if 'bndbox' in json_data['annotation']['object']: # len(object)=1
            print('bbox 1개')
            box_id +=1
            image_id=id       
            id= box_id
            
            obj_name = json_data['annotation']['object']['name']
            print(obj_name)
            # 수정사항
            if obj_name == 'pen(pencil)':
                obj_name ='pen'
            category_id=category_list[obj_name]
            
            bbox=[
                int(json_data['annotation']['object']['bndbox']['xmin']),
                int(json_data['annotation']['object']['bndbox']['ymin']),
                int(json_data['annotation']['size']['width']),
                int(json_data['annotation']['size']['height'])
            ] # x,y,width,height
            sub = {"area":area, "iscrowd":iscrowd, "image_id":image_id, "bbox":bbox, "category_id":category_id, "id":id, "ignore":ignore, "segmentation":segmentation}
            total_sub.append(sub)
            
        else: 
            print('bbox 2개이상')
            for sub_data in json_data['annotation']['object']:
                box_id +=1
                image_id=id       
                id= box_id

                obj_name = sub_data['name']
                print(obj_name)
                # 수정사항
                if obj_name == 'pen(pencil)':
                    obj_name ='pen'
                
                bbox= [
                    int(sub_data['bndbox']['xmin']),
                    int(sub_data['bndbox']['ymin']),
                    int(json_data['annotation']['size']['width']),
                    int(json_data['annotation']['size']['height'])
                ] # x,y,width,height     
                sub = {"area":area, "iscrowd":iscrowd, "image_id":image_id, "bbox":bbox, "category_id":category_id, "id":id, "ignore":ignore, "segmentation":segmentation}
                total_sub.append(sub)
```
