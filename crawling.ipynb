```python
from selenium import webdriver
from urllib.parse import quote_plus
from urllib.request import urlopen
import os
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import re

def save_images(images, save_path):
    for index, image in enumerate(images[:999]):  # images[:크롤링하고 싶은 사진 개수]
        src = image.get_attribute('src')
#         print('src: ',src)
        t = urlopen(src).read()
        ex = re.findall('^\S+\.(\S+)&type',src)
        if ex:
#             print('ex', ex)
            ex=ex[0]
            if ex == 'gif':
                continue
            file = open(os.path.join(save_path, str(index + 1) + "."+ str(ex)), "wb")
            file.write(t)
            print("img save " + save_path + str(index + 1) + "." + str(ex))
        else:
            print('pass')

def create_folder_if_not_exists(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)
 
 
def make_url(search_term):
    # 네이버 이미지 검색
    base_url = 'https://search.naver.com/search.naver?where=image&section=image&query='
    # CCL 상업적 이용 가능 옵션
    end_url = '&res_fr=0&res_to=0&sm=tab_opt&color=&ccl=2' \
              '&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&recent=0&datetype=0&startdate=0&enddate=0&gif=0&optStr=&nso_open=1'
    return base_url + quote_plus(search_term) + end_url
 
def crawl_images(search_term,add_height):
    # URL 생성
    url = make_url(search_term)
 
    # chrome 브라우저 열기
#     driver = webdriver.Chrome('chromedriver')
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.implicitly_wait(3)  # 브라우저를 오픈할 때 시간간격을 준다.
    driver.get(url)
    time.sleep(1.5)
    driver.execute_script('window.scrollTo(0,450)')
    
    # 전체 스크롤 크기
    now_height = driver.execute_script("return window.pageYOffset")
    print('start_ = ',now_height)
    
    # 커서 내리기
    val = 0
    while 1:
        driver.execute_script('window.scrollTo(0,{})'.format(now_height+add_height))
        time.sleep(1.5)
        
        end_loc = driver.execute_script('return document.documentElement.scrollHeight')
        now_height = driver.execute_script("return window.pageYOffset")
        print('now/end= {} / {}'.format(now_height,end_loc))
        
        if now_height==val:
            break
        else:
            val = now_height
            
    # 이미지 긁어오기
    images = driver.find_elements_by_class_name("_image")
 
    # 저장 경로 설정
    save_path = "C:/Users/dongjae/[datacampus]/sample_images"
    create_folder_if_not_exists(save_path)
 
    # 이미지 저장
    save_images(images, save_path)
 
    # 마무리
    print(search_term + " 저장 성공")
    driver.close()
```

### 안내
셀 실행 후 -> 윈도우창 전체화면

내리는 간격은 화면 크기, 검색어별로 상이

스크롤 간격마다 겹치는 부분이 존재해야 안정적으로 다운로드 가능해요

```python
if __name__ == '__main__':
    crawl_images(input('원하는 검색어: '),
                 int(input('내리는 간격:')))
```
